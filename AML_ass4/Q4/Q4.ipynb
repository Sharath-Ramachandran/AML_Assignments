```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
df= pd.read_csv('iris.data', header =None)
```

```python
df.head(10)
```

```python
target_names= df[4].unique()
```

```python
colors = ['r','g','b']
lw = 1
```

```python
target_names
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df[4])
df[4]=le.transform(df[4])
```

```python
pca = PCA(n_components=2)
n= df.values
X= n[:,:4]
y= n[:,4]
Xr= pca.fit(X).transform(X)
```

```python
pca.explained_variance_ratio_
```

```python
pca.components_
```

```python
x_axis=Xr[:,0]
y_axis=Xr[:,1]
```

```python
#plt.scatter(x_axis,y_axis)
for color, i, target_name in zip(colors, [0, 1, 2], ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']):
    plt.scatter(x_axis[y == i], y_axis[y == i], color=color, alpha=.7, lw=lw,
                label=target_names[i])
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset by first and second component')
plt.show()
```

```python
from sklearn.manifold import TSNE
X_embedded = TSNE(n_components=2).fit_transform(X)
x_axis=X_embedded[:,0]
y_axis=X_embedded[:,1]
for color, i, target_name in zip(colors, [0, 1, 2], ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']):
    plt.scatter(x_axis[y == i], y_axis[y == i], color=color, alpha=.7, lw=lw,
                label=target_names[i])
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset by first and second component')
plt.show()
```

```python
from sklearn.datasets.samples_generator import make_swiss_roll
X, label = make_swiss_roll(1500,0.05)
X[:, 1] *= .5
```

```python
pca = PCA(n_components=2)
Xswiss= pca.fit(X).transform(X)
x_axis=Xswiss[:,0]
y_axis=Xswiss[:,1]
plt.scatter(x_axis,y_axis,c=label,cmap='rainbow')
```
